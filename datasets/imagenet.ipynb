```python
import os

def load_class_mapping(mapping_file='./map_clsloc.txt'):
    class_mapping = {}
    with open(mapping_file, 'r') as f:
        for line in f:
            class_id, _, class_name = line.strip().split(' ', 2)  # Split on the first two spaces
            class_mapping[class_id] = class_name
    return class_mapping

def rename_folders(base_dir, class_mapping):
    for class_id in os.listdir(base_dir):
        class_dir = os.path.join(base_dir, class_id)
        if os.path.isdir(class_dir):
            # Get the new class name based on the class ID
            new_class_name = class_mapping.get(class_id, None)
            if new_class_name:
                new_class_dir = os.path.join(base_dir, new_class_name)
                os.rename(class_dir, new_class_dir)  # Rename the folder
                print(f"Renamed {class_id} to {new_class_name}")

# Load the class mapping from map_clsloc.txt
class_mapping = load_class_mapping()

# Directory where your folders are located
base_dir = './data/imagenet_val/ILSVRC2012'

# Rename the folders based on the mapping
rename_folders(base_dir, class_mapping)
```

```python
import os
import shutil

def load_class_mapping(file_path):
    class_mapping = {}
    with open(file_path, 'r') as f:
        for line in f:
            class_id, _, class_name = line.strip().split(maxsplit=2)
            class_mapping[class_id] = class_name
    return class_mapping

def rename_folders(base_dir, class_mapping):
    for class_id, new_class_name in class_mapping.items():
        class_dir = os.path.join(base_dir, class_id)
        new_class_dir = os.path.join(base_dir, new_class_name)
        
        # Check if the new class name folder already exists, and remove it if so
        if os.path.exists(new_class_dir):
            shutil.rmtree(new_class_dir)  # Delete the existing folder
        
        # Rename the folder
        if os.path.exists(class_dir):
            os.rename(class_dir, new_class_dir)
            print(f"Renamed {class_id} to {new_class_name}")
        else:
            print(f"Folder {class_id} not found!")

# Load the class mapping from the file
map_file_path = './map_clsloc.txt'
class_mapping = load_class_mapping(map_file_path)

base_dir = './data/imagenet_val/ILSVRC2012'
rename_folders(base_dir, class_mapping)
```
