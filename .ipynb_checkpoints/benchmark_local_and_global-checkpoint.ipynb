```python
data_type = 'ImageNet'
model_name = 'densenet121'
xai_method = 'Grad-CAM'
gpu_name = "RTX-2080-Ti"

# new GPU/CPU's TDP (Thermal Design Power) should be added in 'utils/benchmark_utils.py'  
# check your GPU/CPU by running 'utils/check_hardware.ipynb'

# number of images to get average efficiency statistics from
num_images = 100 

#---------------------------------------------------------------------------
#  available datasets:    CIFAR10, STL10, ImageNet, URL
#  available ML models:   resnet50, convnext-t, efficientnet-b0, swin-t, regnet-y-8gf, mobilenet-v3-large, densenet121
#  available GPUs:        GTX-1050, RTX-2080-Ti, P100, K80, Tesla-T4
#  available CPUs:        Intel-Core-i7-9700K, AMD-Ryzen-5-3600, Intel-Xeon-E5-2650-v3, default-CPU
#  Gradient-based XAI:    CAM, Grad_CAM, Grad_CAM_plus_plus, GradientShap, GuidedBackprop, DeepLiftShap, InputXGradient, Saliency, IntegratedGradients
#  Perturbation-based:    LIME, PermutationImportance, AblationStudies, RISE, Morris_Sensitivity_Analysis

#  Global methods (requires multiple similar images):   GlobalShap, GlobalDeepLift
```

```python
#  Define the class to focus on for global methods (e.g., 0 for 'airplane' in CIFAR-10)
selected_class = 0 

#  List any global methods here
is_global_method = xai_method in ['GlobalShap','GlobalDeepLift']
```

```python
#  fill the URL links if data_type = 'URL'
image_urls = [
    'https://cdn-boasclserv.pressidium.com/wp-content/uploads/2021/02/Carpets.png',
    'https://static01.nyt.com/images/2012/07/06/sports/SOCCER/SOCCER-superJumbo.jpg',
] if data_type == 'URL' else None
```

```python
# Set the number of images to visualize
if data_type == 'URL':
    num_visualizations = num_images   # will visualize all URL images 
elif data_type == 'CIFAR10':
    num_visualizations = 1
else:
    num_visualizations = 3
```

## Data and Model Loading

```python
model = load_model(model_name=model_name, device=device)
```

```python
# Select the appropriate transform based on the method
if xai_method == 'Grad-CAM' or xai_method == 'GlobalShap' or xai_method == 'GlobalDeepLift':
    transform = transform_model_input
else:
    transform = transform_no_resize # Other methods does not require resizing
```

```python
data_source, labels = load_dataset(
    data_type=data_type,
    num_images=num_images,
    image_urls=image_urls,
    selected_class=selected_class,
    transform=transform,
    is_global_method=is_global_method
)
```

## Dynamic XAI Method Import

```python
# Base directory containing the subdirectories with XAI methods
base_dir = "xai_methods"
sub_dirs = ["Local_Perturbation", "Local_Backpropagation", "Global_Perturbation", "Global_Backpropagation"]

# Function to dynamically import the specified XAI method
def load_xai_method(xai_method):
    for sub_dir in sub_dirs:
        method_path = os.path.join(base_dir, sub_dir, f"{xai_method}.py")  # Construct full path to module file
        if os.path.isfile(method_path):  # Check if the file exists
            # Load module from the specified path
            spec = importlib.util.spec_from_file_location(xai_method, method_path)
            method_module = importlib.util.module_from_spec(spec)
            spec.loader.exec_module(method_module)
            print(f"Successfully loaded {xai_method} from {sub_dir} based methods")
            return method_module
    raise ModuleNotFoundError(f"{xai_method} not found in any of the specified directories.")

# Load the XAI method module
method_module = load_xai_method(xai_method)

# Retrieve necessary functions from the XAI method module
generate_attribution = method_module.generate_attribution
measure_avg_time_across_images = method_module.measure_avg_time_across_images

# Optional visualization function (if exists in the module)
visualize_attribution = getattr(method_module, 'visualize_attribution', None)
visualize_aggregated_attribution = getattr(method_module, 'visualize_aggregated_attribution', None)
```

## Warm-Up for Stability

```python
# Optional warm-up
if hasattr(method_module, 'warm_up'):
    method_module.warm_up(model)
```

# Main Run for Time, Memory and Visualizations

```python
# Process batch of images based on the dataset type
if is_global_method:
    # For GlobalShap or any future global method, process a batch of images
    if data_type == 'URL':
        # URL data only has images, no labels
        avg_time_taken, times, representative_image, aggregated_attribution = measure_avg_time_across_images(
            data_source, model, generate_attribution, url_data=True
        )
    else:
        avg_time_taken, times, representative_image, aggregated_attribution = measure_avg_time_across_images(
            data_source, model, generate_attribution
        )
else:
    # For local methods, process images one by one
    avg_time_taken, times = measure_avg_time_across_images(data_source, model, generate_attribution)

print(f"Average time taken across {len(data_source)} images: {avg_time_taken:.4f} seconds")

peak_memories = []
net_memory_changes = []
cached_visualizations = []

# Measure memory usage for each image, caching a few attributions for visualization
for idx, data in enumerate(data_source):
    image, label = data if isinstance(data, (tuple, list)) else (data, None)
    image = image.to(device)

    # Get predicted class
    with torch.no_grad():
        output = model(image)
        _, predicted_class = torch.max(output, 1)

    # Measure memory with generate_attribution
    peak_memory, net_memory_change = measure_memory(image, predicted_class, generate_attribution, model)
    peak_memories.append(peak_memory)
    net_memory_changes.append(net_memory_change)

    # Cache a few attributions for visualization (done separately to not affect time and memory calculations)
    if idx < num_visualizations:
        attribution = generate_attribution(image, predicted_class, model)
        cached_visualizations.append((image, attribution, label))

    torch.cuda.empty_cache()  # Clear cache to avoid memory accumulation
```

```python
# Set up directory for visualizations if it doesn't exist
visualization_dir = "experiment_results/visualizations"
os.makedirs(visualization_dir, exist_ok=True)

# Check if the method is global or local
if is_global_method:
    # For global methods, visualize the aggregated attribution only once
    visualization_filename = f"{visualization_dir}/{xai_method}_{model_name}_{data_type}_{gpu_name}_global.png"
    visualize_aggregated_attribution(aggregated_attribution, save_path=visualization_filename)
else:
    # For local methods, visualize each cached attribution individually
    if visualize_attribution:
        for i, (image, attribution, label) in enumerate(cached_visualizations):
            # Define the filename with a structured name for each individual visualization
            visualization_filename = f"{visualization_dir}/{xai_method}_{model_name}_{data_type}_{gpu_name}_img{i+1}.png"
            
            # Save the visualization using visualize_attribution (this will show it once per image)
            visualize_attribution(image, attribution, label, labels, model, model_name, save_path=visualization_filename)
```

```python
# If global method, visualize overlay
if is_global_method:
    # Overlay SHAP attribution on a representative image if URL or CIFAR/STL
    plt.figure(figsize=(8, 8))
    plt.imshow(representative_image.transpose(1, 2, 0))  # Display the representative image
    plt.imshow(aggregated_attribution.squeeze(), cmap='inferno', alpha=0.8)  # Overlay heatmap with transparency
    plt.colorbar()
    plt.title("Global SHAP Attribution on Representative Image")
    plt.show()
```

## Statistical Summary

```python
# Calculate averages and standard deviations
avg_peak_memory = sum(peak_memories) / len(peak_memories)
avg_net_memory_change = sum(net_memory_changes) / len(net_memory_changes)
std_time = statistics.stdev(times)
std_peak_memory = statistics.stdev(peak_memories)
std_net_memory_change = statistics.stdev(net_memory_changes)

# Calculate energy consumption based on average time and GPU name
energy_consumed_kWh, used_gpu_name = calculate_energy(avg_time_taken, gpu_name, xai_method, model_name)

# Print out the results
print("----------------------------------------------------------------------------")
print(f"Average time taken: {avg_time_taken:.4f} seconds (±{std_time:.4f})")
print(f"Average peak memory used: {avg_peak_memory:.4f} MiB (±{std_peak_memory:.4f})")
print(f"Average net memory change: {avg_net_memory_change:.4f} MiB (±{std_net_memory_change:.4f})")
print(f"Energy consumed: {energy_consumed_kWh:.4f} kWh")
```

## Save experimental results

```python
# Example calculations for averages and standard deviations
avg_values = {
    "time": avg_time_taken,
    "peak_memory": avg_peak_memory,
    "net_memory_change": avg_net_memory_change
}

std_devs = {
    "time": std_time,
    "peak_memory": std_peak_memory,
    "net_memory_change": std_net_memory_change
}

# Create the summary dictionary
summary = {
    "energy_consumed_kWh": energy_consumed_kWh,
    "gpu_name": gpu_name,
    "num_images": num_images
}

# Create per-image results as before
per_image_results = [
    {"image_index": i, "time": time, "peak_memory": peak_mem, "net_memory_change": net_mem}
    for i, (time, peak_mem, net_mem) in enumerate(zip(times, peak_memories, net_memory_changes))
]

# Save the main experiment results
save_experiment_results(xai_method, model_name, data_type, summary, per_image_results, avg_values, std_devs)
```
