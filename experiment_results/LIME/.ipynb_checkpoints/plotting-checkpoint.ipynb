```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import json

# Load the JSONL file
file_path = "summary.jsonl"  # Adjust path if needed
with open(file_path, "r") as f:
    data = [json.loads(line) for line in f]

df = pd.DataFrame(data)

# Define efficiency metrics and parameter names
metrics = ["average_time", "average_peak_memory", "average_net_memory_change", "energy_consumed_kWh"]
parameters = ["n_segments", "n_samples", "compactness", "sigma"]

# Function to create a summary table for each parameter
def generate_summary_tables(df, parameters, metrics):
    summary_tables = {}
    for param in parameters:
        grouped = df.groupby(param)[metrics].mean().round(4)
        summary_tables[param] = grouped
    return summary_tables

# Generate tables
tables = generate_summary_tables(df, parameters, metrics)

# Display tables (can export as CSV if needed)
for param, table in tables.items():
    print(f"\nEffect of {param} on Efficiency Metrics:\n")
    display(table)

# Plotting
sns.set(style="whitegrid", font_scale=1.1)
for param in parameters:
    fig, axs = plt.subplots(2, 2, figsize=(12, 8))
    fig.suptitle(f"Effect of {param} on Efficiency Metrics", fontsize=16)

    for i, metric in enumerate(metrics):
        ax = axs[i // 2, i % 2]
        sns.lineplot(data=df, x=param, y=metric, marker="o", ax=ax)
        ax.set_title(metric.replace("_", " ").title())
        ax.set_xlabel(param)
        ax.set_ylabel(metric)

    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()
```
